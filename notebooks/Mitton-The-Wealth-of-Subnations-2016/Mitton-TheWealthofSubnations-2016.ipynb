# 1. Introduction

This paper studies the impact of geography and institutions on the divergent fortunes of subnational regions within countries throughout the world. There is an extensive literature that studies this relationship, finding that geography and institutions are key for economic development. However, existing studies tend to focus on cross-country differences leaving aside within-country variation, which is no less important. In fact, data of GDP per capita across the globe reveals important differences in the levels of economic development of countries' subnational regions. For instance, the state of Nuevo León in Mexico exceeds that of Chiapas by a factor of four, while in the Russian Federation, Nents Oblast is wealthier than Ingushetia Republic by a factor of 69. 

The aim of this paper is thus to analyze the effect of geography and institutions on subnational development, employing an exhaustive set of geographic variables and subnational institutional measures for a large set of countries. 

The main results show that geography plays a significant role in shapping within-country differences in development. In contrast, the correlation between institutions and income is usually negative. Additional tests suggest that the lack of a positive within-country effect of institutions on development could be explained by the dominance of national over subnational institutions. In particular, using various proxies for autonomy and federalism, estimations reveal a significant positive relation between institutions and development in regions that have greater autonomy, suggesting that strong regional institutions may aid development when not dominanted by national institutions.

These results contribute to the overall debate on geography vs. institutions by showing that geography has a significant impact on development even after controlling for country-level factors, as well as by showing the conditions under which subnational institutions matter for regional development.

The following bullets provide additional details on the data used to construct the main variables of the analysis, as well as the coding used to import the raw data and get it ready to perform the corresponding estimations.

* Economic development variable (main dependent variable): data on GDP per-capita for 2005 at the subnational level come from various data sources. The most common sources are the national statistics bureaus of the individual countries (contains data for 35 countries), Eurostat (data for 28 European Nations) and the United Nations Development Program (data for 30 nations). Data for the remaining eight countries come from different data sources. The following code, imports GDP data and constructs GDP per capita estimates for 2005 adjusted for purchasing power parity using the World Bank PPP adjustment factors.

```python
# Import GDP data
dataGDP = ["GDP Data", "Country level data"]
GDP, country = [pd.read_stata(filepath_or_buffer= pathout + "{}.dta".format(i)) for i in dataGDP]
```

```python
## Prepare GDP data for estimates
# Merge subnational data with country level data
GDP = pd.merge(GDP, country, on="countrycode", how="left", validate="m:1")
# Adjust GDP per-capita estimates for purchasing power parity
GDP['gdppc_ppp']=GDP['gdppc2005']/GDP['pppfactor']
GDP.loc[GDP.gdpcurrency == "PPP", 'gdppc_ppp'] = GDP.gdppc2005
# .80412 is the EUR/USD exchange rate in 2005
GDP.loc[GDP.gdpcurrency != GDP.officialcurrency2011, 'gdppc_ppp'] = GDP.gdppc2005*GDP.xrate*(1/0.80412)/GDP.pppfactor
country = GDP.groupby(['countrycode'])['population2005'].sum()
GDP = pd.merge(GDP, country, on="countrycode", how="left", validate="m:1")
GDP['popratio']=GDP['population2005_x']/GDP['population2005_y']
GDP['popratioXgdpppp']=GDP['popratio']*GDP['gdppc_ppp']
country = GDP.groupby(['countrycode'])['popratioXgdpppp'].sum()
GDP = pd.merge(GDP, country, on="countrycode", how="left", validate="m:1")
GDP['gdppc_ppp_rel']=GDP['gdppc_ppp']/GDP['popratioXgdpppp_y']
GDP['gdppc_rel_wb']=GDP['gdppc_ppp_rel']/GDP['gdppcppp_wb']
GDP = GDP.sort_values(by=['isocode'],ignore_index=True)
GDP1=GDP
```

* Geographical variables (explanatory variables): to construct geographic variables the author employs a number of geospatial datasets maintained by governmental and private organizations (Google Earth, National Oceanic Atmospheric Administration, Esri, Climate Research Unit at the University of East Anglia, Mineral Resources Data System, among others). These sources offer detailed global mappings of a wide variety of general geographic characteristics (e.g., elevation or terrain ruggedness), climate (e.g., precipitation or temperature), and natural resources (e.g., oil or iron). The following code imports geographic, climate and natural resources data, and prepares it to perform estimates.

```python
# Import geographic data
dataGEO = ["missingiso", "elv", "Map Data","Water Data", "Google Earth Data", "Tropical Storm Data", "Fault Data", "Ruggedness Data", "malaria data", "Ocean Access Data"]
missingiso, elv, mapdata, water, google, tropical, fault, rugged, malaria, ocean = [pd.read_stata(filepath_or_buffer= pathout + "{}.dta".format(i)) for i in dataGEO]
```

```python
## Merge geographic data with GDP data and prepare it for estimates
missingiso = pd.merge(missingiso, elv, on=["latitude", "longitude"], how="left", validate="m:1")
missingiso = missingiso.groupby(['isocode'])['elv'].mean()
GDP = pd.merge(GDP, missingiso, on="isocode", how="left", validate="1:1")
GDP = pd.merge(GDP, mapdata, on="isocode", how="left", validate="1:1")
GDP = pd.merge(GDP, google, on="isocode", how="left", validate="1:1")
tropical = tropical.groupby(['isocode'])['objectid'].count()
GDP = pd.merge(GDP, tropical, on="isocode", how="left", validate="1:1")
GDP = GDP.rename(columns={'objectid':'stormcount'})
fault1 = fault.groupby(['isocode'])['fault_leng_km'].count()
GDP = pd.merge(GDP, fault1, on="isocode", how="left", validate="1:1")
GDP = GDP.rename(columns={'fault_leng_km':'faultcount'})
fault2 = fault.groupby(['isocode'])['fault_leng_km'].sum()
GDP = pd.merge(GDP, fault2, on="isocode", how="left", validate="1:1")
GDP = GDP.rename(columns={'fault_leng_km':'faultlength'})
GDP = pd.merge(GDP, rugged, on="isocode", how="left", validate="1:1")
GDP['cclat']=GDP['cclatdeg']+GDP['cclatmin']/100+GDP['cclatsec']/10000
GDP = pd.merge(GDP, malaria, on="isocode", how="left", validate="1:1")
GDP = pd.merge(GDP, ocean, on="index", how="left", validate="1:1")
GDP['latitude']=np.abs(GDP['cclat'])
GDP['loggdppc_rel_wb']=np.abs(GDP['gdppc_rel_wb'])
GDP.loc[(GDP['stormcount'].isnull()) & (GDP['isocode']!="PKTA"),'stormcount']=0
GDP.loc[(GDP['faultcount'].isnull()) & (GDP['isocode']!="PKTA"),'faultcount']=0
GDP.loc[(GDP['faultlength'].isnull()) & (GDP['isocode']!="PKTA"),'faultlength']=0
GDP['stormcount_sqkm']=GDP['stormcount']/GDP['totalareasqkm']
GDP['faultcount_sqkm']=GDP['faultcount']/GDP['totalareasqkm']
GDP['faultlength_sqkm']=GDP['faultlength']/GDP['totalareasqkm']
# Rescale variables
GDP['tri']=GDP['tri']/100000
GDP['slope']=GDP['slope']/1000
GDP['logtri']=np.log(1+GDP['tri'])
GDP['logstormcount']=np.log(1+GDP['stormcount'])
GDP['logfaultcount']=np.log(1+GDP['faultcount'])
GDP['loggdppc_ppp']=np.log(GDP['gdppc_ppp'])
```

```python
## Merge climate data with GDP data and prepare it for estimates
missingiso = pd.merge(missingiso, pre, on=["latitude", "longitude"], how="left")
missingiso = missingiso.sort_values(by=["latitude", "longitude"],ignore_index=True)
missingiso = pd.merge(missingiso, rd, on=["latitude", "longitude"], how="left")
missingiso = pd.merge(missingiso, tmp, on=["latitude", "longitude"], how="left")
missingiso = pd.merge(missingiso, dtr, on=["latitude", "longitude"], how="left")
missingiso = pd.merge(missingiso, reh, on=["latitude", "longitude"], how="left")
missingiso = pd.merge(missingiso, sunp, on=["latitude", "longitude"], how="left")
missingiso = pd.merge(missingiso, frs, on=["latitude", "longitude"], how="left")
missingiso = pd.merge(missingiso, wnd, on=["latitude", "longitude"], how="left")
missingiso = pd.merge(missingiso, elv, on=["latitude", "longitude"], how="left")
pre = missingiso.loc[: , "pre1":"pre12"]
missingiso['pre_avg'] = pre.mean(axis=1)
rd = missingiso.loc[: , "rd1":"rd12"]
missingiso['rd_avg'] = rd.mean(axis=1)
tmp = missingiso.loc[: , "tmp1":"tmp12"]
missingiso['tmp_avg'] = tmp.mean(axis=1)
dtr = missingiso.loc[: , "dtr1":"dtr12"]
missingiso['dtr_avg'] = dtr.mean(axis=1)
reh = missingiso.loc[: , "reh1":"reh12"]
missingiso['reh_avg'] = reh.mean(axis=1)
sunp = missingiso.loc[: , "sunp1":"sunp12"]
missingiso['sunp_avg'] = sunp.mean(axis=1)
frs = missingiso.loc[: , "frs1":"frs12"]
missingiso['frs_avg'] = frs.mean(axis=1)
wnd = missingiso.loc[: , "wind1":"wind12"]
missingiso['wnd_avg'] = wnd.mean(axis=1)
missingiso = missingiso.groupby(['isocode']).mean()
missingiso = missingiso.reset_index()
clicols = ["pre","rd","tmp","dtr","reh","sunp","frs","wnd"]
cols = [col+'_avg' for col in clicols]
missingiso = missingiso[["isocode"] + cols]
GDP = pd.merge(GDP, missingiso, on="isocode", how="left", validate="1:1")
# Construct variable of percentage of land area that falls in tropical or subtropical biome
biome1 = biome[(biome['fid_biomes'] > 10) & (biome['fid_biomes']<15) & (biome['fid_biomes'].notnull())]
biome1 = biome1.rename(columns={'biome_area':'trop_area'})
biome1 = biome1.groupby(['isocode'])['trop_area'].sum()
biome2 = biome[(biome['fid_biomes'].notnull())] 
biome2 = biome2.rename(columns={'biome_area':'all_area'})
biome2 = biome2.groupby(['isocode'])['all_area'].sum()
biome2 = pd.merge(biome2, biome1, on="isocode", how="left", validate="1:1")
biome2['trop_pct']=biome2['trop_area']/biome2['all_area']
biome2.loc[(biome2['trop_pct'].isnull()),'trop_pct']=0
GDP = pd.merge(GDP, biome2, on="isocode", how="left", validate="1:1")
```

* Institutional variables (explanatory variables): to construct subnational institutional variables, the author employs data from several different organizations that survey companies and individuals. The surveys, in total encompassing over 170,000 respondents from 86 countries, each identify the subnational region of respondents, allowing to create subnational-level measures of institutional quality. From each survey, the author extracts questions referring to the quality of institutions. To make the scale of questions comparable within surveys, the log of responses is taken from questions that have a maximun response and the negative of responses is taken for questions in which a higher number indicates worse institutions. Finally, each question is standardized so that all questions are weighted equally when aggregated. Extracted questions are classified in four categories of institutions: property rights protection, corruption control, law and order, and regulatory efficiency. The following code imports institutions data and prepares it to perform estimates.

```python
# Import institutions data
dataINS = ["institutions concordance","afrodata","TAFdata","qogdata","WBESdata","dbdata","lapopdata","latinobarometro"]
inst,afro,taf,qog,wbes,db,lapop,latino= [pd.read_stata(filepath_or_buffer= pathout + "{}.dta".format(i),convert_categoricals=False) for i in dataINS]
```

```python
## Merge institutions data with GDP data and prepare it for estimates
GDP2 = pd.merge(GDP1, inst, on="index",how="left",validate="1:1")
GDP2 = GDP2.sort_values(by=['index_afro'],ignore_index=True)
GDP2 = pd.merge(GDP2, afro, on="index_afro", how="left", validate="m:m")
GDP2 = GDP2.sort_values(by=['index_taf'],ignore_index=True)
taf = taf.sort_values(by=['index_taf'],ignore_index=True)
GDP2 = pd.merge(GDP2, taf, on="index_taf", how="left", validate="m:m")
GDP2 = GDP2.sort_values(by=['index_qog'],ignore_index=True)
qog = qog.sort_values(by=['index_qog'],ignore_index=True)
GDP2 = pd.merge(GDP2, qog, on="index_qog", how="left", validate="m:m")
GDP2 = GDP2.sort_values(by=['index_wbes'],ignore_index=True)
wbes = wbes.sort_values(by=['index_wbes'],ignore_index=True)
GDP2 = pd.merge(GDP2, wbes, on="index_wbes", how="left", validate="m:m")
GDP2 = GDP2.sort_values(by=['index_db'],ignore_index=True)
db = db.sort_values(by=['index_db'],ignore_index=True)
GDP2 = pd.merge(GDP2, db, on="index_db", how="left", validate="m:m")
GDP2 = GDP2.sort_values(by=['index_lapop'],ignore_index=True)
lapop = lapop.sort_values(by=['index_lapop'],ignore_index=True)
GDP2 = pd.merge(GDP2, lapop, on="index_lapop", how="left", validate="m:m")
GDP2 = GDP2.sort_values(by=['index_lbo'],ignore_index=True)
latino = latino.sort_values(by=['index_lbo'],ignore_index=True)
GDP2 = pd.merge(GDP2, latino, on="index_lbo", how="left", validate="m:m")
columns = ["afro", "taf", "wbes", "db", "lapop", "lbo"]
pr = GDP2[['pr_'+col for col in columns]]
GDP2['pr_all'] = pr.mean(axis=1)
re = GDP2[['re_'+col for col in columns]]
GDP2['re_all'] = re.mean(axis=1)
columns = ["afro", "taf", "wbes", "qog", "lapop", "lbo"]
lo = GDP2[['lo_'+col for col in columns]]
GDP2['lo_all'] = lo.mean(axis=1)
cc = GDP2[['cc_'+col for col in columns]]
GDP2['cc_all'] = cc.mean(axis=1)
comp2 = GDP2[['comp2_afro','comp2_taf','comp2_wbes','comp2_db','comp2_qog','comp2_lbo','comp2_lapop']]
GDP2['comp2_all'] = comp2.mean(axis=1)
inst = GDP2.loc[:,['pr_all','cc_all','lo_all','re_all']]
GDP2['inst_all'] = inst.mean(axis=1)
columns = ["afro", "wbes", "db", "lapop", "lbo"]
prpos = GDP2[['prpos_'+col for col in columns]]
GDP2['prpos_all'] = prpos.mean(axis=1)
repos = GDP2[['repos_'+col for col in columns]]
GDP2['repos_all'] = repos.mean(axis=1)
columns = ["afro", "wbes", "qog", "lapop", "lbo"]
lopos = GDP2[['lopos_'+col for col in columns]]
GDP2['lopos_all'] = lopos.mean(axis=1)
ccpos = GDP2[['ccpos_'+col for col in columns]]
GDP2['ccpos_all'] = ccpos.mean(axis=1)
instpos = GDP2.loc[:,['prpos_all','ccpos_all','lopos_all','repos_all']]
GDP2['instpos_all'] = instpos.mean(axis=1)
GDP2 = GDP2.sort_values(by=['isocode'],ignore_index=True)
GDP2 = GDP2.loc[:,['isocode','pr_all','cc_all','lo_all','re_all','inst_all','comp2_afro','comp2_taf','comp2_wbes','comp2_db','comp2_qog','comp2_lbo','comp2_lapop','comp2_all',
                  'prpos_all','repos_all','lopos_all','ccpos_all','instpos_all']]
GDP = pd.merge(GDP, GDP2, on='isocode', how='left', validate="1:1")
```

# 3. Empirical approach

To investigate the determinants of economic development at the subnational level using the data described in the previous section, this paper estimates regression equations of the following form:

$logy_{ci}=\alpha+X'_{ci}\beta+\delta_{c}+\varepsilon_{ci}$ (1)

where $y_{ci}$ is GDP per capita for subnational region $i$ in country $c$, $X_{ci}$ is a vector containing one or more proposed determinants of development in subnational region $i$ in country $c$, and $\delta_{c}$ represents a full set of country fixed effects. Standard errors are clustered at the country level.

# 5. Conclusions

This paper finds a significant role of geographic features in explaining within-country divergence in economic development. Conversely, results reveal no clear positive effect of institutional quality on income level of subnational regions. Furthermore, the author finds that the lack of a positive within-country effect on institutions may be due to the fact that subnational institutions are dominanted by national institutions. Consistent with this interpretation, additional regressions reveal a positive relation between subnational institutions and development among regions that have greater autonomy.

# 6. Additional analysis Colombian context

This last section evaluates whether the role of geographical features in explaining within-country variation in income is also relevant when using data for Colombian municipalities.

```python
# Import data for Colombia
datosCOL = ["caracteristicasg", "conflicto", "gobierno", "temp", "rain", "Censo2018", "TMI"]
caracteristicasg, conflicto, gobierno, temp, rain, censo2018,TMI = [pd.read_stata(filepath_or_buffer= pathout_col + "{}.dta".format(i)) for i in datosCOL]
```

```python
# Keep data for 2005
caracteristicasg['ano'] = pd.to_datetime(caracteristicasg['ano'])
conflicto['ano'] = pd.to_datetime(conflicto['ano'])
caracteristicasg_2005 = caracteristicasg[caracteristicasg['ano'].dt.year == 2005]
conflicto_2005 = conflicto[conflicto['ano'].dt.year == 2005]
gobierno_2005 = gobierno[gobierno['ano']== 2005]
```

```python
# Merge data of economic development and climate 
dfCOL = pd.merge(caracteristicasg_2005, gobierno_2005, on='codmpio')
dfCOL = pd.merge(dfCOL, conflicto_2005, on='codmpio')
dfCOL = pd.merge(dfCOL, rain, how='left', on='codmpio')
dfCOL = pd.merge(dfCOL, temp, how='left', on='codmpio')
dfCOL = pd.merge(dfCOL, censo2018, how='left', on='codmpio')
dfCOL = pd.merge(dfCOL, TMI, how='left', on='codmpio')
```

```python
# Create additional variables
dfCOL['logPIBpc'] = np.log(dfCOL['PIBpc2018'])
dfCOL['PIBpcmill'] = dfCOL['PIBpc2018']/1000000
dfCOL['altura2'] = np.log(1+dfCOL['altura'])
dfCOL['meanrain2'] = dfCOL['meanrain']/100
```

## 6.1. Descriptive statistics

The following descriptive statistics show that there is a positive correlation of GDP per capita in Colombian municipalities with elevation. In contrast, higher levels of precipitation and temperature are related with lower subnational incomes. Futhermore, maps reveal the presence of central-periphery patterns in the incidence of these variables. In particular, lower levels of GDP per-capita and geographical disadvantages appear to be concentrated in the Colombian peripheral regions: Pacific, Caribbean and Amazonic.

```python
# Function to draw scatter plots
def ScatterPlot(dfin, var0='x', var1='y', labelvar='indcode', 
                    xlabel='variable x', 
                    ylabel='variable y',
                    filename='file.pdf'):
    '''
    Plot the association between var0 and var in dataframe using labelvar for labels. 
    '''
    sns.set(rc={'figure.figsize':(11.7,8.27)})
    sns.set_context("talk")
    df = dfin.copy()
    df = df.dropna(subset=[var0, var1]).reset_index(drop=True)
    # Plot
    fig, ax = plt.subplots()
    sns.regplot(x=var0, y=var1, data=df, ax=ax)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    plt.savefig(pathgraphs + filename, dpi=300, bbox_inches='tight')
    pass
```
